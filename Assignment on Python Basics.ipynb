5. What is mutability in Python?

- Mutability in Python refers to the ability of an object to be modified after it is created.


6. Why are lists mutable, but tuples are immutable?

7. What is the difference between “==” and “is” operators in Python?

- The == operator checks for value equality, meaning it compares whether the values of two objects are the same. The is operator checks for identity equality, meaning it determines whether two objects refer to the same memory location.


8. What are logical operators in Python?

9. What is type casting in Python?

- Type casting in Python is the process of converting one data type into another, such as converting an integer to a string or a float to an integer.


10. What is the difference between implicit and explicit type casting?

- Implicit type casting is automatically done by Python when it converts one data type to another, usually to avoid data loss example: converting an integer to a float when necessary.

- Explicit type casting requires the programmer to manually convert one data type to another using functions like int(), float(), or str().


11. What is the purpose of conditional statements in Python?

- Conditional statements in Python allow the program to execute specific code based on whether a condition is true or false, enabling decision-making in the program.

12. How does the elif statement work?

- The elif statement in Python allows you to check multiple conditions. If the if condition is false, Python evaluates the elif condition, and if it's true, it executes the corresponding block of code.

13. What is the difference between for and while loops?

- A for loop iterates over a sequence (like a list or range) for a specific number of times, while a while loop continues executing as long as a given condition is true.









14. Describe a scenario where a while loop is more suitable than a for loop?

- A while loop is more suitable when you don’t know the number of iterations in advance and want to continue looping until a specific condition is met, such as waiting for user input or monitoring a changing value.

6. Write a function to demonstrate the behavior of mutable and immutable arguments.                 

```python
def modify_arguments(mutable_arg, immutable_arg):
    mutable_arg.append(10)

    immutable_arg += 5

    print("Inside function:")
    print(f"Mutable argument (list): {mutable_arg}")
    print(f"Immutable argument (integer): {immutable_arg}")

my_list = [1, 2, 3]
my_number = 20

print("Before function call:")
print(f"Mutable argument (list): {my_list}")
print(f"Immutable argument (integer): {my_number}")

modify_arguments(my_list, my_number)

print("\nAfter function call:")
print(f"Mutable argument (list): {my_list}")
print(f"Immutable argument (integer): {my_number}")
```

9. Write a Python program to convert user input from string to integer, float, and boolean types.


```python
user_input = input("Enter anything: ")

# Convert to integer
if user_input.isdigit():
    int_value = int(user_input)
    print("Converted to integer:", int_value)
else:
    print("Could not convert to integer.")

# Convert to float
if user_input.replace('.', '', 1).isdigit() and user_input.count('.') < 2:
    float_value = float(user_input)
    print("Converted to float:", float_value)
else:
    print("Could not convert to float.")

# Convert to boolean
bool_value = bool(user_input)
print("Converted to boolean:", bool_value)
```

13. Write a Python program to find the sum of all even numbers between 1 and 50.


```python

sum_of_evens = 0

for number in range(1, 51):
    if number % 2 == 0:
        sum_of_evens += number
print("The sum of all even numbers between 1 and 50 is:", sum_of_evens)
```

14. Write a program to reverse a string using a while loop.


```python

input_string = input("Enter a string to reverse: ")

reversed_string = ""
index = len(input_string) - 1

while index >= 0:
    reversed_string += input_string[index]
    index -= 1

print("Reversed string:", reversed_string)
```

15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.


```python
number = int(input("Enter a number to calculate its factorial: "))

factorial = 1
i = number

while i > 0:
    factorial *= i
    i -= 1
    i -= 1

print(f"The factorial of {number} is: {factorial}")
```
