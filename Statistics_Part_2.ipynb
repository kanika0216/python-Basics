                                                 **Theory**

1. *What is hypothesis testing in statistics?*  
   Hypothesis testing is a statistical method to test assumptions about a population parameter using sample data.  

2. *What is the null hypothesis, and how does it differ from the alternative hypothesis?*  
   The null hypothesis (H₀) assumes no effect or difference, while the alternative hypothesis (H₁) suggests a significant effect or difference.  

3. *What is the significance level in hypothesis testing, and why is it important?*  
   The significance level (α) is the probability of rejecting a true null hypothesis, typically set at 0.05 (5%).  

4. *What does a P-value represent in hypothesis testing?*  
   The P-value represents the probability of obtaining the observed results if the null hypothesis is true.  

5. *How do you interpret the P-value in hypothesis testing?*  
   A small P-value (≤ α) suggests rejecting H₀; a large P-value (> α) suggests failing to reject H₀.  

6. *What are Type 1 and Type 2 errors in hypothesis testing?*  
   Type 1 error (α) is rejecting a true H₀; Type 2 error (β) is failing to reject a false H₀.  

7. *What is the difference between a one-tailed and a two-tailed test in hypothesis testing?*  
   A one-tailed test checks for a directional effect, while a two-tailed test checks for any difference.  

8. *What is the Z-test, and when is it used in hypothesis testing?*  
   A Z-test is used to compare population means when the sample size is large (n ≥ 30) and variance is known.  

9. *How do you calculate the Z-score, and what does it represent in hypothesis testing?*  
   The Z-score measures how many standard deviations a value is from the mean; Z = (X - μ) / σ.  

10. *What is the T-distribution, and when should it be used instead of the normal distribution?*  
    The T-distribution is used for small sample sizes (n < 30) when the population variance is unknown.  

11. *What is the difference between a Z-test and a T-test?*  
    A Z-test is for large samples with known variance, while a T-test is for small samples with unknown variance.  

12. *What is the T-test, and how is it used in hypothesis testing?*  
    A T-test compares the means of two groups to determine if they are significantly different.  

13. *What is the relationship between the Z-test and T-test in hypothesis testing?*  
    Both compare means, but a T-test adjusts for smaller samples and unknown variance.  

14. *What is a confidence interval, and how is it used to interpret statistical results?*  
    A confidence interval is a range of values estimating a population parameter with a given confidence level (e.g., 95%).  

15. *What is the margin of error, and how does it affect the confidence interval?*  
    The margin of error defines the range within which the true population parameter is expected to lie, affecting the precision of the confidence interval.  

16. *How is Bayes' Theorem used in statistics, and what is its significance?*  
    Bayes' Theorem updates probabilities based on new evidence and is crucial in Bayesian statistics.  

17. *What is the Chi-square distribution, and when is it used?*  
    The Chi-square distribution is used for categorical data analysis and hypothesis testing for variance and independence.  

18. *What is the Chi-square goodness of fit test, and how is it applied?*  
    The Chi-square goodness of fit test checks if observed data matches an expected distribution.  

19. *What is the F-distribution, and when is it used in hypothesis testing?*  
    The F-distribution is used in variance comparison tests like ANOVA and F-tests.  

20. *What is an ANOVA test, and what are its assumptions?*  
    ANOVA analyzes variance among multiple groups and assumes normality, independence, and equal variance.  

21. *What are the different types of ANOVA tests?*  
    One-way ANOVA (one factor), Two-way ANOVA (two factors), and Repeated Measures ANOVA.  

22. *What is the F-test, and how does it relate to hypothesis testing?*  
    The F-test compares variances of two or more groups to determine if they are significantly different.

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
import numpy as np
from scipy.stats import norm

sample = [65, 70, 75, 80, 85]
population_mean = 75
population_std_dev = 10

sample_mean = np.mean(sample)
sample_size = len(sample)

z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between sample and population mean.")
```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

sample = [62, 65, 70, 68, 72]
population_mean = 70
population_std_dev = 5

z_score = (np.mean(sample) - population_mean) / (population_std_dev / np.sqrt(len(sample)))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

x = np.linspace(-4, 4, 1000)
critical_value = 1.96
y = norm.pdf(x)

plt.plot(x, y, label="Standard Normal Distribution")
plt.axvline(critical_value, color='red', linestyle='--', label="Critical Region (±1.96)")
plt.axvline(-critical_value, color='red', linestyle='--')
plt.axvline(z_score, color='blue', linestyle='-', label=f"Z-Score ({z_score:.2f})")
plt.legend()
plt.title("Two-Tailed Z-Test Decision Region")
plt.xlabel("Z-Score")
plt.ylabel("Density")
plt.show()

print("Z-Score:", z_score)
print("P-Value:", p_value)
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

def type1_type2_errors(alpha, true_mean, sample_mean, std_dev, n):
    z_critical = norm.ppf(1 - alpha / 2)
    x = np.linspace(-4, 4, 1000)
    y_null = norm.pdf(x, loc=0, scale=1)
    y_alt = norm.pdf(x, loc=(sample_mean - true_mean) / (std_dev / np.sqrt(n)), scale=1)

    plt.plot(x, y_null, label="Null Hypothesis (H₀)", color='blue')
    plt.plot(x, y_alt, label="Alternative Hypothesis (H₁)", color='green')
    plt.fill_between(x, y_null, where=(x <= -z_critical) | (x >= z_critical), color='blue', alpha=0.3, label="Type I Error (α)")
    plt.fill_between(x, y_alt, where=(x > -z_critical) & (x < z_critical), color='green', alpha=0.3, label="Type II Error (β)")
    plt.axvline(-z_critical, color='red', linestyle='--', label=f"Critical Value (-{z_critical:.2f})")
    plt.axvline(z_critical, color='red', linestyle='--', label=f"Critical Value (+{z_critical:.2f})")
    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Z-Score")
    plt.ylabel("Density")
    plt.legend()
    plt.show()

type1_type2_errors(alpha=0.05, true_mean=0, sample_mean=1, std_dev=1, n=30)
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence_level):
    mean = np.mean(data)
    std_error = np.std(data, ddof=1) / np.sqrt(len(data))
    z_score = norm.ppf(1 - (1 - confidence_level) / 2)
    margin_of_error = z_score * std_error
    return mean - margin_of_error, mean + margin_of_error

data = [20, 22, 19, 24, 21, 23]
interval = confidence_interval(data, 0.95)
print("Confidence Interval:", interval)
```

10. Write a Python program to calculate the margin of error for a given confidence level using sample data

```python
import numpy as np
from scipy.stats import norm

def margin_of_error(data, confidence_level):
    std_error = np.std(data, ddof=1) / np.sqrt(len(data))
    z_score = norm.ppf(1 - (1 - confidence_level) / 2)
    return z_score * std_error

data = [20, 22, 19, 24, 21, 23]
error = margin_of_error(data, 0.95)
print("Margin of Error:", error)
```

12. Perform a Chi-square test for independence between two categorical variables in Python.

13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data

```python
import numpy as np

def calculate_expected(observed):
    row_totals = observed.sum(axis=1)
    col_totals = observed.sum(axis=0)
    grand_total = observed.sum()
    expected = np.outer(row_totals, col_totals) / grand_total
    return expected

observed_data = np.array([[30, 20], [50, 40]])
expected_frequencies = calculate_expected(observed_data)
print("Expected Frequencies:\n", expected_frequencies)
```

20.  Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.



```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt

data = {
    'Factor1': np.tile(['A', 'B'], 15),
    'Factor2': np.repeat(['X', 'Y', 'Z'], 10),
    'Response': np.random.normal(loc=10, scale=2, size=30)
}

df = pd.DataFrame(data)

model = ols('Response ~ C(Factor1) * C(Factor2)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)

means = df.groupby(['Factor1', 'Factor2']).mean().unstack()
means.plot(kind='bar', figsize=(8, 6))
plt.title('Interaction Plot')
plt.ylabel('Mean Response')
plt.show()
```

24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results .

```python
import numpy as np
from scipy.stats import chi2

data = [10, 12, 9, 14, 11, 13, 12, 10, 15, 9]
sample_variance = np.var(data, ddof=1)
population_variance = 4
n = len(data)

chi2_stat = (n - 1) * sample_variance / population_variance
p_value = 2 * min(1 - chi2.cdf(chi2_stat, df=n-1), chi2.cdf(chi2_stat, df=n-1))

print("Chi-Square Statistic:", chi2_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis: The population variance is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in the population variance.")
```

26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.



```python
import numpy as np
from scipy.stats import f
import matplotlib.pyplot as plt

data1 = np.random.normal(loc=20, scale=5, size=30)
data2 = np.random.normal(loc=22, scale=7, size=30)

var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

f_stat = var1 / var2 if var1 > var2 else var2 / var1
df1 = len(data1) - 1
df2 = len(data2) - 1

p_value = 1 - f.cdf(f_stat, df1, df2)

x = np.linspace(0, 5, 1000)
y = f.pdf(x, df1, df2)

plt.plot(x, y, label="F-Distribution", color="blue")
plt.axvline(f_stat, color="red", linestyle="--", label=f"F-Statistic ({f_stat:.2f})")
plt.title("F-Test for Variance Comparison")
plt.xlabel("F-Statistic")
plt.ylabel("Density")
plt.legend()
plt.show()

print("F-Statistic:", f_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the variances.")
```
