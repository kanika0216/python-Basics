1. What is a parameter?

 A parameter is a variable used by a model that is learned during the training phase (e.g., weights in linear regression).

2. What is correlation?

 Correlation measures the relationship between two variables, indicating whether they move together or inversely.

3. What does negative correlation mean?

 Negative correlation means that as one variable increases, the other decreases (e.g., rainfall and outdoor activities).

4. Define Machine Learning. What are the main components in Machine Learning?

 Machine Learning enables systems to learn from data to make predictions; its main components include data, features, models, training, and evaluation.

5. How does loss value help in determining whether the model is good or not?

 Loss value quantifies the error between predicted and actual values; lower loss indicates a better model.

6. What are continuous and categorical variables?

 Continuous variables have numerical values (e.g., height); categorical variables represent categories (e.g., colors).

7. How do we handle categorical variables in Machine Learning? What are the common techniques?

 Categorical variables are handled using encoding techniques like one-hot encoding or label encoding.

8. What do you mean by training and testing a dataset?

 Training is used to fit the model, while testing evaluates its performance on unseen data.

9. What is sklearn.preprocessing??

 sklearn.preprocessing is a module in Scikit-learn for transforming data, such as scaling and encoding.

10. What is a Test set?

 A test set is a subset of data reserved for evaluating the model's performance.

11. How do we split data for model fitting (training and testing) in Python?

 Use train_test_split from Scikit-learn to split the dataset into training and testing sets.

12. How do you approach a Machine Learning problem?

 Steps include defining the problem, collecting data, performing EDA, feature engineering, selecting a model, and evaluation.

13. Why do we have to perform EDA before fitting a model to the data?

 EDA helps understand data patterns, detect outliers, and identify relationships for better model building.

14. what is correlation?

 Correlation is a statistical measure that describes how two variables are related.

 It indicates an inverse relationship where one variable decreases as the other increases.

16. How can you find correlation between variables in Python?

 Use df.corr() for correlation matrices or scipy.stats.pearsonr for specific variable pairs.

17. What is causation? Explain difference between correlation and causation with an example.

 Causation implies one variable affects another (e.g., smoking causes lung issues); correlation (e.g., ice cream sales and drowning) does not imply causation.

  An optimizer adjusts model weights to minimize loss; examples include Gradient Descent, Adam, and RMSprop.

 It’s a Scikit-learn module for linear modeling, such as Linear Regression and Logistic Regression.

 Feature scaling standardizes data ranges for consistent model performance, especially with distance-based algorithms.

24. How do we perform scaling in Python?
